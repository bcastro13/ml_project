```python
df5 = pd.merge(df3, df4, on="id")
df5.head()
```

```python
df1.drop(
    ["Unnamed: 0", "count", "hate_speech", "offensive_language", "neither"],
    axis=1,
    inplace=True,
)
df1.head()
```

```python
df6 = df5.append(df2)
df6.head()
```

```python
print(len(df6))
df6 = df6[df6["toxic"] != -1]
print(len(df6))
df6.head()
```

```python
conditions = [
    df6["identity_hate"] == 1,
    (df6["toxic"] == 1) | (df6["obscene"] == 1),
    (df6["toxic"] != 1) & (df6["obscene"] != 1) & (df6["identity_hate"] != 1),
]
values = [0, 1, 2]
df6["class"] = np.select(conditions, values)
df6.head()
```

```python
df6.drop(
    ["id", "toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"],
    axis=1,
    inplace=True,
)
df6.head()
```

```python
df1.rename(columns={"tweet": "text"}, inplace=True)
df6.rename(columns={"comment_text": "text"}, inplace=True)
```

```python
final_df = df1.append(df6)
final_df
```
