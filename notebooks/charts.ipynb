```python
names = []
training_acc = []
validation_acc = []
validation_err = []
for key in metrics.keys():
    names.append(key)
    training_acc.append(metrics[key]["training_acc"])
    validation_acc.append(metrics[key]["validation_acc"])
    validation_err.append(metrics[key]["validation_err"])
```

```python
plt.rcParams["figure.figsize"] = (15, 10)
plt.bar(names, training_acc, color="y")
plt.xticks(rotation=70)
for i in range(len(training_acc)):
    plt.annotate(
        str(round(training_acc[i], 3)),
        xy=(names[i], training_acc[i]),
        ha="center",
        va="bottom",
        weight="bold",
    )
plt.xlabel("Models")
plt.ylabel("Training Accuracy")
plt.title("Model Training Accuracy")


plt.show()
```

```python
plt.bar(names, validation_acc, color="y")
plt.xticks(rotation=70)
for i in range(len(validation_acc)):
    plt.annotate(
        str(round(validation_acc[i], 3)),
        xy=(names[i], validation_acc[i]),
        ha="center",
        va="bottom",
        weight="bold",
    )
plt.xlabel("Models")
plt.ylabel("Validation Accuracy")
plt.title("Model Validation Accuracy")

plt.show()
```

```python
plt.bar(names, validation_err, color="y")
plt.xticks(rotation=70)
for i in range(len(validation_err)):
    plt.annotate(
        str(round(validation_err[i], 3)),
        xy=(names[i], validation_err[i]),
        ha="center",
        va="bottom",
        weight="bold",
    )
plt.xlabel("Models")
plt.ylabel("Validation Error")
plt.title("Model Validation Error")

plt.show()
```

```python
df = pd.DataFrame()
df["Models"] = names
df["Training Accuracy"] = training_acc
df["Validation Accuracy"] = validation_acc
```
